```python
train = train.drop(train.index[1273:1821])
print(train)
```

```python
#train['healthy'][i] es 0 o 1
for i in range(len(train)):
  train['healthy'][i]
```

```python
#Train
path = '/content/images/'
lista = os.listdir(path)   # con esto se puede manejar una carpeta como una lista :)
lista = sorted(lista)      # ordenamos la lista, pues listdir() las deja desordenadas.
for i in tqdm(range(1273), desc = 'procesando'):  #tqdm es para ver el % de procesado del for i in, desc es el print
  img = cv.imread(path + lista[i])
  if train['healthy'][i] == 1:
    cv.imwrite("/content/images/Train/Healthy/" + lista[i], img)  #imwrite(donde queremos guardar, nombre que queremos)
  else: 
    cv.imwrite("/content/images/Train/Unhealthy/" + lista[i], img)
```

```python
#Test
path = '/content/images/'
lista = os.listdir(path)   # con esto se puede manejar una carpeta como una lista :)
lista = sorted(lista)      # ordenamos la lista, pues listdir() las deja desordenadas.
for i in tqdm(range(1273,1821), desc = 'procesando'):  #tqdm es para ver el % de procesado del for i in, desc es el print
  img = cv.imread(path + lista[i])
  if test['healthy'][i] == 1:
    cv.imwrite("/content/images/Test/Healthy/" + lista[i], img)  #imwrite(donde queremos guardar, nombre que queremos)
  else: 
    cv.imwrite("/content/images/Test/Unhealthy/" + lista[i], img)
```
